```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_html('Trade_Map_-_List_of_exporters_for_the_selected_product_in_2022_(Iron_and_steel).xls')
df=df[2]
df.columns=df.iloc[1,:]
df=df.iloc[2:,:]
df=df.set_index('Exporters')
df['Exports in 2022']=[float(i) for i in df.iloc[:,0]]
iron=df['Exports in 2022']
iron=iron.sort_values(ascending=False)
iron
```

# GDP

```python
dfg=pd.read_excel("API_NY.GDP.MKTP.CD_DS2_en_excel_v2_5795872.xls")
dfg.columns=dfg.iloc[2,:]
dfg=dfg.set_index('Country Name')
dfg.head()
```

```python
dfg1=dfg.transpose().iloc[3:,3:]
dfg1['Year']=[int(i) for i in dfg1.index]
dfg1=dfg1.set_index('Year')
dfg1.columns=list(dfg1.columns)
dfg1.head()
```

```python
gdp=dfg1.transpose()
gdp=gdp[2022]
gdp
```

```python
y=list();z=list();g=list();ex=list()
for p in gdp.index:
    for i in list(iron.index):
        if p==i:
            y.append(p)
            z.append(i) 
            g.append(gdp[p])
            ex.append(iron[i]*1000)      # convert thousand USD to Current USD

dnew=pd.DataFrame([y,z,g,ex]).transpose().iloc[:167,:]
dnew.columns=['Country_gdp','Country_iron','GDP','Exports of Iron & Steel']
```

```python
# Verify if merging is well done
dnew[dnew['Country_gdp']=='China'],(gdp['China'],iron['China'])
```

```python
dnew=dnew.sort_values('Exports of Iron & Steel',ascending=False)
dnew=dnew.iloc[-166:,:]             # remove Total related to "World"
dnew.head(20)
```

```python

import matplotlib.pyplot as plt
fig=plt.subplot()
plt.scatter(x=list(dnew.GDP)[0:50],y=list(dnew['Exports of Iron & Steel'])[0:50])
plt.show()
```
