```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_html('Trade_Map_-_List_of_exporters_for_the_selected_product_in_2022_(Iron_and_steel).xls')
df=df[2]
df.columns=df.iloc[1,:]
df=df.iloc[2:,:]
df=df.set_index('Exporters')
df['Exports in 2022']=[float(i) for i in df.iloc[:,0]]
iron=df['Exports in 2022']
iron=iron.sort_values(ascending=False)
iron
```

```python
iron1=iron.sort_values(ascending=True)
iron1=iron1[0:(len(iron1)-1)]
iron1=iron1[-20::]
fig=plt.subplot()
plt.barh(iron1.index,iron1.values/1000000)
plt.title('Top 20 Exporters of Iron and Steel in 2022')
plt.ylabel('Countries')
plt.xlabel('Millions of USD')
for i in range(len(iron1)):
    plt.text(y=iron1.index[i],x=(iron1.values[i]/1000000)+0.25,s=str(round(iron1.values[i]/1000000,1)),color='black')
# plt.text(x=iron1.index,y=iron1.values,s=iron1.values,color='black')
fig.axes.get_xaxis().set_ticks([])
plt.text(x=-6,y=-5,s='Source: Realized by Raulin L. Cadet, with data from TradeMap.')
fig.spines[['right', 'top','bottom']].set_visible(False) 
plt.xticks(visible=False)
```
